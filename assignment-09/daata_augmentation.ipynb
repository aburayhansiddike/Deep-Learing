```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
from tqdm import tqdm
import time

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# CIFAR-10 class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']

# Define transforms - WITHOUT augmentation
transform_original = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
])

# Define transforms - WITH augmentation
transform_augmented = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.RandomCrop(32, padding=4),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
])

# Test transform (no augmentation for testing)
transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
])

# Load datasets
print("Loading CIFAR-10 dataset...")

# Original dataset (no augmentation)
trainset_original = torchvision.datasets.CIFAR10(root='./data', train=True, 
                                                 download=True, transform=transform_original)
train_loader_original = DataLoader(trainset_original, batch_size=128, shuffle=True, num_workers=2)

# Augmented dataset
trainset_augmented = torchvision.datasets.CIFAR10(root='./data', train=True, 
                                                  download=False, transform=transform_augmented)
train_loader_augmented = DataLoader(trainset_augmented, batch_size=128, shuffle=True, num_workers=2)

# Test dataset
testset = torchvision.datasets.CIFAR10(root='./data', train=False, 
                                      download=False, transform=transform_test)
test_loader = DataLoader(testset, batch_size=128, shuffle=False, num_workers=2)

print(f"Training samples: {len(trainset_original)}")
print(f"Test samples: {len(testset)}")

# Define CNN Architecture
class CIFAR10CNN(nn.Module):
    def __init__(self, num_classes=10):
        super(CIFAR10CNN, self).__init__()
        
        # First convolutional block
        self.conv_block1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Dropout(0.25)
        )
        
        # Second convolutional block
        self.conv_block2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Dropout(0.25)
        )
        
        # Third convolutional block
        self.conv_block3 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Dropout(0.25)
        )
        
        # Classifier
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 4 * 4, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )
        
    def forward(self, x):
        x = self.conv_block1(x)
        x = self.conv_block2(x)
        x = self.conv_block3(x)
        x = self.classifier(x)
        return x

# Function to show augmented images
def show_augmented_samples():
    """Display original vs augmented images"""
    # Get some samples without augmentation
    original_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, 
                                                    transform=transforms.ToTensor())
    
    # Create augmentation transform for display
    aug_transform = transforms.Compose([
        transforms.RandomHorizontalFlip(p=1.0),  # Always apply for demo
        transforms.RandomRotation(15),
        transforms.RandomCrop(32, padding=4),
        transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1)
    ])
    
    fig, axes = plt.subplots(2, 8, figsize=(16, 4))
    fig.suptitle('Original vs Augmented Images', fontsize=16, fontweight='bold')
    
    for i in range(8):
        # Get original image
        img, label = original_dataset[i]
        
        # Apply augmentation
        img_aug = aug_transform(img)
        
        # Show original
        axes[0, i].imshow(np.transpose(img, (1, 2, 0)))
        axes[0, i].set_title(f'Original\n{class_names[label]}', fontsize=10)
        axes[0, i].axis('off')
        
        # Show augmented
        axes[1, i].imshow(np.transpose(img_aug, (1, 2, 0)))
        axes[1, i].set_title(f'Augmented\n{class_names[label]}', fontsize=10)
        axes[1, i].axis('off')
    
    plt.tight_layout()
    plt.show()

# Training function
def train_model(model, train_loader, test_loader, epochs=50, model_name="Model"):
    """Train the model and return training history"""
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=15, gamma=0.5)
    
    train_losses = []
    train_accuracies = []
    test_accuracies = []
    
    print(f"\n{'='*50}")
    print(f"Training {model_name}")
    print(f"{'='*50}")
    
    for epoch in range(epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0
        
        pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{epochs}')
        for batch_idx, (data, targets) in enumerate(pbar):
            data, targets = data.to(device), targets.to(device)
            
            optimizer.zero_grad()
            outputs = model(data)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total_train += targets.size(0)
            correct_train += predicted.eq(targets).sum().item()
            
            pbar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{100.*correct_train/total_train:.2f}%'
            })
        
        # Calculate epoch metrics
        epoch_loss = running_loss / len(train_loader)
        train_acc = 100. * correct_train / total_train
        
        # Testing phase
        model.eval()
        correct_test = 0
        total_test = 0
        with torch.no_grad():
            for data, targets in test_loader:
                data, targets = data.to(device), targets.to(device)
                outputs = model(data)
                _, predicted = outputs.max(1)
                total_test += targets.size(0)
                correct_test += predicted.eq(targets).sum().item()
        
        test_acc = 100. * correct_test / total_test
        
        # Store metrics
        train_losses.append(epoch_loss)
        train_accuracies.append(train_acc)
        test_accuracies.append(test_acc)
        
        scheduler.step()
        
        print(f'Epoch [{epoch+1}/{epochs}] - Loss: {epoch_loss:.4f}, Train Acc: {train_acc:.2f}%, Test Acc: {test_acc:.2f}%')
    
    return train_losses, train_accuracies, test_accuracies

# Function to evaluate model and create confusion matrix
def evaluate_model(model, test_loader, model_name):
    """Evaluate model and return predictions for confusion matrix"""
    model.eval()
    all_predictions = []
    all_targets = []
    correct = 0
    total = 0
    
    with torch.no_grad():
        for data, targets in test_loader:
            data, targets = data.to(device), targets.to(device)
            outputs = model(data)
            _, predicted = outputs.max(1)
            
            all_predictions.extend(predicted.cpu().numpy())
            all_targets.extend(targets.cpu().numpy())
            
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
    
    accuracy = 100. * correct / total
    print(f'{model_name} Final Test Accuracy: {accuracy:.2f}%')
    
    return np.array(all_predictions), np.array(all_targets), accuracy

# Show augmented samples first
print("Displaying augmented samples...")
show_augmented_samples()

# Initialize models
print("\nInitializing models...")
model_original = CIFAR10CNN().to(device)
model_augmented = CIFAR10CNN().to(device)

# Train model without augmentation
print("\n" + "="*60)
print("TRAINING MODEL WITHOUT DATA AUGMENTATION")
print("="*60)
start_time = time.time()
train_losses_orig, train_acc_orig, test_acc_orig = train_model(
    model_original, train_loader_original, test_loader, epochs=25, 
    model_name="CNN without Augmentation"
)
time_orig = time.time() - start_time

# Train model with augmentation
print("\n" + "="*60)
print("TRAINING MODEL WITH DATA AUGMENTATION")
print("="*60)
start_time = time.time()
train_losses_aug, train_acc_aug, test_acc_aug = train_model(
    model_augmented, train_loader_augmented, test_loader, epochs=25,
    model_name="CNN with Augmentation"
)
time_aug = time.time() - start_time

# Evaluate both models
print("\n" + "="*60)
print("EVALUATION RESULTS")
print("="*60)
pred_orig, targets_orig, acc_orig = evaluate_model(model_original, test_loader, "Original Model")
pred_aug, targets_aug, acc_aug = evaluate_model(model_augmented, test_loader, "Augmented Model")

# Create comprehensive comparison plots
fig = plt.figure(figsize=(20, 12))

# Training Loss Comparison
plt.subplot(2, 3, 1)
epochs_range = range(1, len(train_losses_orig) + 1)
plt.plot(epochs_range, train_losses_orig, 'b-', label='Without Augmentation', linewidth=2)
plt.plot(epochs_range, train_losses_aug, 'r-', label='With Augmentation', linewidth=2)
plt.title('Training Loss Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# Training Accuracy Comparison
plt.subplot(2, 3, 2)
plt.plot(epochs_range, train_acc_orig, 'b-', label='Without Augmentation', linewidth=2)
plt.plot(epochs_range, train_acc_aug, 'r-', label='With Augmentation', linewidth=2)
plt.title('Training Accuracy Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True, alpha=0.3)

# Test Accuracy Comparison
plt.subplot(2, 3, 3)
plt.plot(epochs_range, test_acc_orig, 'b-', label='Without Augmentation', linewidth=2)
plt.plot(epochs_range, test_acc_aug, 'r-', label='With Augmentation', linewidth=2)
plt.title('Test Accuracy Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True, alpha=0.3)

# Confusion Matrix - Original Model
plt.subplot(2, 3, 4)
cm_orig = confusion_matrix(targets_orig, pred_orig)
sns.heatmap(cm_orig, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names)
plt.title(f'Confusion Matrix - Original\nAccuracy: {acc_orig:.2f}%', fontsize=12, fontweight='bold')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Confusion Matrix - Augmented Model
plt.subplot(2, 3, 5)
cm_aug = confusion_matrix(targets_aug, pred_aug)
sns.heatmap(cm_aug, annot=True, fmt='d', cmap='Reds', 
            xticklabels=class_names, yticklabels=class_names)
plt.title(f'Confusion Matrix - Augmented\nAccuracy: {acc_aug:.2f}%', fontsize=12, fontweight='bold')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Performance Summary Bar Chart
plt.subplot(2, 3, 6)
models = ['Without\nAugmentation', 'With\nAugmentation']
accuracies = [acc_orig, acc_aug]
colors = ['skyblue', 'lightcoral']
bars = plt.bar(models, accuracies, color=colors, alpha=0.8, edgecolor='black', linewidth=1)

# Add accuracy values on bars
for bar, acc in zip(bars, accuracies):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5, f'{acc:.2f}%',
             ha='center', va='bottom', fontweight='bold', fontsize=12)

plt.title('Final Test Accuracy Comparison', fontsize=14, fontweight='bold')
plt.ylabel('Accuracy (%)')
plt.ylim(0, 100)
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# Print detailed comparison
print("\n" + "="*60)
print("DETAILED COMPARISON RESULTS")
print("="*60)
print(f"{'Metric':<25} {'Without Aug':<15} {'With Aug':<15} {'Improvement':<15}")
print("-" * 70)
print(f"{'Final Test Accuracy':<25} {acc_orig:<15.2f} {acc_aug:<15.2f} {acc_aug-acc_orig:<15.2f}")
print(f"{'Training Time (sec)':<25} {time_orig:<15.1f} {time_aug:<15.1f} {time_aug-time_orig:<15.1f}")
print(f"{'Best Test Accuracy':<25} {max(test_acc_orig):<15.2f} {max(test_acc_aug):<15.2f} {max(test_acc_aug)-max(test_acc_orig):<15.2f}")

# Calculate per-class accuracy
def calculate_per_class_accuracy(predictions, targets):
    """Calculate per-class accuracy"""
    per_class_acc = {}
    for i, class_name in enumerate(class_names):
        class_mask = targets == i
        if class_mask.sum() > 0:
            class_correct = (predictions[class_mask] == targets[class_mask]).sum()
            per_class_acc[class_name] = 100.0 * class_correct / class_mask.sum()
        else:
            per_class_acc[class_name] = 0.0
    return per_class_acc

per_class_orig = calculate_per_class_accuracy(pred_orig, targets_orig)
per_class_aug = calculate_per_class_accuracy(pred_aug, targets_aug)

# Plot per-class accuracy comparison
plt.figure(figsize=(12, 6))
x = np.arange(len(class_names))
width = 0.35

plt.bar(x - width/2, [per_class_orig[name] for name in class_names], 
        width, label='Without Augmentation', color='skyblue', alpha=0.8)
plt.bar(x + width/2, [per_class_aug[name] for name in class_names], 
        width, label='With Augmentation', color='lightcoral', alpha=0.8)

plt.xlabel('Classes', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.title('Per-Class Accuracy Comparison', fontsize=14, fontweight='bold')
plt.xticks(x, class_names, rotation=45)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print("\nPer-Class Accuracy Comparison:")
print(f"{'Class':<12} {'Without Aug':<12} {'With Aug':<12} {'Improvement':<12}")
print("-" * 50)
for class_name in class_names:
    improvement = per_class_aug[class_name] - per_class_orig[class_name]
    print(f"{class_name:<12} {per_class_orig[class_name]:<12.2f} {per_class_aug[class_name]:<12.2f} {improvement:<12.2f}")

print(f"\nOverall improvement: {acc_aug - acc_orig:.2f}% accuracy increase with data augmentation")
print("Analysis Complete!")
```
